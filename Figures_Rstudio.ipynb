```python
%%R
print(getwd())
MAIN = "" # DRIVE DIRECTORY WHERE THE FOLDER "Rstudio_csv" IS 
setwd(MAIN)
getwd()
library(ggplot2)
library(tikzDevice)
library(extrafont)
font_install('fontcm')
fonts()
loadfonts()

r<-c("0.5","1","0.3333333333333333")
r.map <- list("1"="1","0.5"="2","0.3333333333333333"="3")
level_order<-c("UPC","ICB","HI","WKI")
level_map <-list("UPC"="PC","ICB"="CC","HI"="HI","WKI"="WI","PSKM"="PSKM")
#level_order<-c("UPC","ICB","HI","WKI","PSKM")
#plot_list<-c("n_dis","time","error_real_init","n_iter","n_dis","error_real_conver")
plot_list<-c("time","n_iter","n_dis","error_init","error_conver")
#plot_list<-c("error_conver")
#plot_list<-c("error_real_init","error_real_conver")
#plot_list<-c("n_dis")
plot_title<-list("time"="Elapsed time","error_real_init"="Initial SKM error","error_conver"="Converged surrogate error","n_iter"="Number of iterations","n_dis"="Number of computed distances","error_init"="Initial surrogate error","error_real_conver"="Converged SKM error")
s<-40
color<-c("#FF3333", "#FF9933", "#66CC00", "#0066CC", "#FF33FF")
pdf.options(encoding="default")

for (rho in r){
  for (plot in plot_list){
    
df<-read.csv(paste("./Rstudio_csv/plot",plot,"rho_speed",rho,sep = ""))
val <-plot_title[plot][[1]]
meth <- ""#"Algorithm"
tittle<- paste(plot_title[plot][[1]])#,", m =",r.map[rho][[1]])
cex.lab=0.9
p<-ggplot(df,aes(x=reorder(t,factor(method, level = level_order)), y=value),main = NULL)+theme(legend.position = "top", legend.justification = "center")+
  coord_cartesian(ylim = quantile(df$value, c(0.01, 0.96)))+#scale_y_log10()+
  geom_boxplot( aes( fill=factor(method, level = level_order)),outlier.size = -1)+ scale_fill_manual(values=color,labels = level_map)+
  scale_x_discrete(limits=c("1", "2", "4","10"))+
  facet_grid(K~epsilon, scales="free", space="free_x",  labeller= label_bquote(cols = epsilon~"="~.(epsilon),rows = italic("K")~"="~.(K)))+ylab(NULL)+
  xlab(expression(paste("Number of batches since last drift, ",italic("T"))))+theme(strip.text.x = element_text(size = s))+theme(strip.text.y = element_text(size = s))+
  theme(plot.title = element_text(hjust = 0.5))+theme(text = element_text(family = "Times New Roman",size=s))+theme(axis.text=element_text(size=22))+labs(fill=meth,title=tittle)
ggsave(filename = paste("./Rstudio_csv/Graph_",plot,"_rho_speed_",r.map[rho][[1]],".pdf",sep = ""),width = 18,height = 9,device = "pdf")

print(p)
  }}    
```
